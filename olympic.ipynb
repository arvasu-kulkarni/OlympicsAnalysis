```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
r = requests.get("https://en.wikipedia.org/wiki/2024_Summer_Olympics#Sports")
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
print(soup.prettify())
```

```python
links = soup.find_all('a', title=lambda value: value and 'at the 2024 Summer Olympics' in value)
for n, link in enumerate(links):
    print(n, link['href'])
```

```python
sublist = links[782:]
print(sublist[0]['href'])
sublist_str = []
for link in sublist:
    sublist_str.append('https://en.wikipedia.org/'+link['href'])
sublist_str
```

```python
testLink = sublist_str[0]
satyam = requests.get(testLink)
soupPatil = BeautifulSoup(satyam.content, 'html.parser')
print(soupPatil.prettify())
```

```python
import pandas as pd

# URL of the Wikipedia page containing the table
url = "https://en.wikipedia.org/wiki/2024_Summer_Olympics#Sports"

# Read the HTML table into a list of DataFrame objects
table = soupPatil.find_all('table', {'class':'wikitable sortable plainrowheaders jquery-tablesorter'})
medalTable = table[0]
```

```python
df = pd.read_html(str(medalTable))
df[0]
```

```python
def medalToFraction(table):
    table['Gold'] /= table['Gold'].iloc[-1]
    table['Silver'] /= table['Silver'].iloc[-1]
    table['Bronze'] /= table['Bronze'].iloc[-1]
    table['Total'] /= table['Total'].iloc[-1]

medalToFraction(df[0])
```

```python
df[0]
```

```python
def loadSport(link):
    r = requests.get(link)
    soup = BeautifulSoup(r.content, 'html.parser')
    table = soup.find_all('table', {'class':'wikitable sortable plainrowheaders jquery-tablesorter'})
    medalTable = table[0]
    df = pd.read_html(str(medalTable))
    return df[0]
```

```python
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
main = pd.DataFrame(columns=['NOC', 'Gold', 'Silver', 'Bronze', 'Total'])
for link in sublist_str:
    print(link[31:])
    temp_df = loadSport(link)
    medalToFraction(temp_df)
    for row in temp_df.iterrows():
        if 'Nation' in row[1]:
            row[1].rename({'Nation':'NOC'}, inplace=True)
        country = row[1]['NOC']
        if country in main['NOC'].values:
            main.loc[main['NOC'] == country, 'Gold'] += row[1]['Gold']
            main.loc[main['NOC'] == country, 'Silver'] += row[1]['Silver']
            main.loc[main['NOC'] == country, 'Bronze'] += row[1]['Bronze']
            main.loc[main['NOC'] == country, 'Total'] += row[1]['Total']
        else:
            main.loc[len(main)] = row[1]
```

```python
for row in main.iterrows():
    if 'Totals (' in row[1]['NOC']:
        main.drop(row[0], inplace=True)
main
```

```python
main.sort_values(by='Gold', ascending=False, inplace=True)
main.reset_index(drop=True, inplace=True)
main
```

```python
main['NOC' == 'Individual Neutral Athletes'] += main['NOC' == 'Individual Neutral Athletes[A]']
main.drop('Individual Neutral Athletes[A]', inplace=True)
```

```python
main.to_csv('2024_Olympics_Medal_Predictions.csv', index=False)
```
